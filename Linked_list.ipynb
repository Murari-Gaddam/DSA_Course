##**Creating a Node**



```
TASK:
    Creates a Node
    
IDEA:
    Creates a node by assigning a value
    and declare a next node

INPUTS:
    value - value of the node
  
COMPLEXITY:
    Space - O(1)
    Time - O(1)
```

````python
class Node:
  def __init__ (self,value):

    self.value = value
    self.next = None
````

<hr>

##**Creating a Linked List**



```
TASK:
    Creates a linked list

IDEA:
    Initiats a linked list class
    assigns the head and tail to a node

INPUTS:
    value - value of the node

COMPLEXITY:
    Space - O(1)
    Time - O(1)
```



````python
class LinkedList:
  def __init__ (self,value):

    node1 = Node(value)
    self.head = node1
    self.tail = node1
    self.length = 1

list1 = LinkedList(1)
````

<hr>

##**Printing elements of a Linked List**



```
TASK:
    Prints all the elements of a linked list

IDEA:
    Creates a temperary pointer towards head
    and traverses the linked list to print elements

COMPLEXITY:
    Space - O(1)
    Time - O(n)
```

````python
def print_list_func (self):
    temp = self.head
    while temp is not None:
      print(temp.value)
      temp = temp.next

LinkedList.print_list = print_list_func

list1.print_list()
````

<hr>

##**Appending a Node**



```
TASK:
    Appending a node to a linked list

IDEA:
    Adding a node next to the last node
    Making the new node as tail of linked list
    Increment length

INPUTS:
    value - value of the node

COMPLEXITY:
    Space - O(1)
    Time - O(1)
```



````python
def append_func (self , value):
  node2 = Node(value)
  if self.length == 0 :
    self.head = node2
    self.tail = node2
  else:
    self.tail.next = node2
    self.tail = node2
    self.length +=1

LinkedList.append = append_func
list1.append(2)
list1.append(3)
list1.print_list()
````

##**Popping the last Node**



```
TASK:
    Removing the last node of a linked list

IDEA:
    Traversing the list till the 2nd last node
    setting next to None
    Decrement length

RETURN:
    None - if linked list has no node
    temp - returns the whole last node
           value accessed using temp.value

COMPLEXITY:
    Space - O(1)
    Time - O(n)
```



````python
def pop_func (self):
    if self.length == 0:
      return None
    temp = self.head
    pre = self.head

    if self.length == 1:
      self.head = None
      self.tail = None
    else:
      while temp.next:
        pre = temp
        temp = temp.next
      self.tail = pre
      self.tail.next = None
    self.length -=1
    return temp

LinkedList.pop = pop_func
popped = list1.pop()
print(f"the popped value is {popped.value}")
list1.print_list()
````

<hr>

##**Prepending a Node**



```
TASK:
    Prepending a node to a linked list

IDEA:
    Adding a node before the first node
    Making the new node as head of linked list
    Increment length

INPUTS:
    value - value of the node

COMPLEXITY:
    Space - O(1)
    Time - O(1)
```



````python
def prepend_func (self, value):
    node1 = Node(value)
    temp = self.head
    self.head = node1
    self.head.next = temp
    self.length +=1

LinkedList.prepend = prepend_func
list1.prepend(0)
list1.print_list()
````

##**Popping the First Node**

```
TASK:
    Removing the First node of a linked list

IDEA:
    setting temp to removing node
    setting head to temp.next
    setting temp.next to None
    Decrement length

RETURN:
    None - if linked list has no node
    temp - returns the whole First node
           value accessed using temp.value

COMPLEXITY:
    Space - O(1)
    Time - O(1)
```



````python
def pop_first_func (self):
    if self.length == 0:
      return None
    temp = self.head
    if self.length == 1:
      self.head = None
      self.tail = None
    else:
      self.head = temp.next
      temp.next = None
    self.length -=1
    return temp

LinkedList.pop_first = pop_first_func
popped = list1.pop_first()
print(f"the popped value is {popped.value}")
list1.print_list()
````

##**Getting a value at Index**



```
TASK:
    Getting the value at node of given index

IDEA:
    setting temp to head
    traverse the list till index
    return the value at index node

INPUTS:
    index - index of the required node

RETURN:
    None - index negative or greater than
           length of linked list
    temp - returns the whole index node
           value accessed using temp.value

COMPLEXITY:
    Space - O(1)
    Time - O(n)
```



````python
def get_func (self , index):
    temp = self.head
    if index  < 0 or index > self.length  :
      return None
    else:
      for i in range(index):
        temp = temp.next
      return temp

LinkedList.get = get_func
index = int(input("Enter the index"))
val = list1.get(index)
print(f" value at {index} : {val.value}" if val else None)
````
