##**Creating a Node**



```
TASK:
    Creates a Node
    
IDEA:
    Creates a node by assigning a value
    and declare a next node

INPUTS:
    value - value of the node
  
COMPLEXITY:
    Space - O(1)
    Time - O(1)
```

````python
class Node:
  def __init__ (self,value):

    self.value = value
    self.next = None
````

##**Creating a Linked List**



```
TASK:
    Creates a linked list

IDEA:
    Initiats a linked list class
    assigns the head and tail to a node

INPUTS:
    value - value of the node

COMPLEXITY:
    Space - O(1)
    Time - O(1)
```



````python
class LinkedList:
  def __init__ (self,value):

    node1 = Node(value)
    self.head = node1
    self.tail = node1
    self.length = 1

list1 = LinkedList(1)
````
