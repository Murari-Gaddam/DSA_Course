##**Creating a Node**



```
TASK:
    Creates a Node
    
IDEA:
    Creates a node by assigning a value
    and declare a next node

INPUTS:
    value - value of the node
  
COMPLEXITY:
    Space - O(1)
    Time - O(1)
```

````python
class Node:
  def __init__ (self,value):

    self.value = value
    self.next = None
````

<hr>

##**Creating a Linked List**



```
TASK:
    Creates a linked list

IDEA:
    Initiats a linked list class
    assigns the head and tail to a node

INPUTS:
    value - value of the node

COMPLEXITY:
    Space - O(1)
    Time - O(1)
```



````python
class LinkedList:
  def __init__ (self,value):

    node1 = Node(value)
    self.head = node1
    self.tail = node1
    self.length = 1

list1 = LinkedList(1)
````

<hr>

##**Printing elements of a Linked List**



```
TASK:
    Prints all the elements of a linked list

IDEA:
    Creates a temperary pointer towards head
    and traverses the linked list to print elements

COMPLEXITY:
    Space - O(1)
    Time - O(n)
```

````python
def print_list_func (self):
    temp = self.head
    while temp is not None:
      print(temp.value)
      temp = temp.next

LinkedList.print_list = print_list_func

list1.print_list()
````

<hr>

##**Appending a Node**



```
TASK:
    Appending a node to a linked list

IDEA:
    Adding a node next to the last node
    Making the new node as tail of linked list
    Increment length

INPUTS:
    value - value of the node

COMPLEXITY:
    Space - O(1)
    Time - O(1)
```



````python
def append_func (self , value):
  node2 = Node(value)
  if self.length == 0 :
    self.head = node2
    self.tail = node2
  else:
    self.tail.next = node2
    self.tail = node2
    self.length +=1

LinkedList.append = append_func
list1.append(2)
list1.append(3)
list1.print_list()
````
